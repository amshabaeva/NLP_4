```python
import pandas as pd
import gzip

def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield eval(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

df = getDF('All_beauty.json.gz')
```

#### Извлечение именованных сущностей с помощью stanza

Просто чтобы понимать, от чего отталкиваться.

```python
import stanza
stanza.download('en') # download English model
nlp = stanza.Pipeline(lang='en', processors='tokenize,ner') # initialize English neural pipeline
```

Попробуем выделить именнованные сущности в отзывах. Будем выделять сущности только типов 'ORG' и 'PRODUCT'.
Можно заметить, что довольно часто встречается организация Gilette и такие продукты как Brush или Razor. Можно попробовать либо использовать в шаблонах слово расческа, чтобы искать разные расчески, либо слово бритва, чтобы искать бритвы, либо фирму Gillette, чтобы искать разные продукты этого бренда (а может быть и объединить как-то Razor и Gilette). 

```python
for review in list(df2['reviewText']):
    doc = nlp(review)
    for sent in doc.sentences:
        for ent in sent.ents:  # достаем NE
            if ent.type in ['ORG', 'PRODUCT']:
                print(ent)
```

### Оценка  n-грам с помощью разных метрик

Импортируем nltk и все необходимые средства.
Собираем токены всех ревью в один большой список.

```python
import nltk
from nltk.collocations import *
from nltk import ngrams
bigram_measures = nltk.collocations.BigramAssocMeasures()
trigram_measures = nltk.collocations.TrigramAssocMeasures()
fourgram_measures = nltk.collocations.QuadgramAssocMeasures()

all_the_tokens = []
for review in list(df2['reviewText']):
    tokens = nltk.wordpunct_tokenize(review)
    all_the_tokens.extend(tokens)
```

##### Оценка биграм

###### PMI
Собираем биграмы из токенов всех ревью, считаем PMI и оставляем только те биграмы, которые есть в списке grams_2 (Левый сосед + NE)

```python
grams_2_scored_pmi = []
finder = BigramCollocationFinder.from_words(all_the_tokens)
scored = finder.score_ngrams(bigram_measures.pmi)
for bigram in scored:
    if bigram[0] in grams_2:
        grams_2_scored_pmi.append(bigram)

        
sorted(grams_2_scored_pmi, key=lambda x: x[1], reverse=True)
```

###### Likelihood_ratio
Собираем биграмы из токенов всех ревью, считаем likelihood_ratio и оставляем только те биграмы, которые есть в списке grams_2 (Левый сосед + NE)

```python
grams_2_scored_likelihood_ratio = []
finder = BigramCollocationFinder.from_words(all_the_tokens)
scored = finder.score_ngrams(bigram_measures.likelihood_ratio)
for bigram in scored:
    if bigram[0] in grams_2:
        grams_2_scored_likelihood_ratio.append(bigram)

        
sorted(grams_2_scored_likelihood_ratio, key=lambda x: x[1], reverse=True)
```

###### Dice
Собираем биграмы из токенов всех ревью, считаем dice и оставляем только те биграмы, которые есть в списке grams_2 (Левый сосед + NE)

```python
grams_2_scored_dice = []
finder = BigramCollocationFinder.from_words(all_the_tokens)
scored = finder.score_ngrams(bigram_measures.dice)
for bigram in scored:
    if bigram[0] in grams_2:
        grams_2_scored_dice.append(bigram)

        
sorted(grams_2_scored_dice, key=lambda x: x[1], reverse=True)
```
